```python
# Initialize city and time the build
t0 = time.perf_counter()
city = City(dimensions=(22, 22))

# Small helper to add many buildings from a declarative list
def _add_many(city, entries):
    """Accepts a list of dicts where each dict has:\n    - 'type' (building_type), 'door' (tuple) and either 'blocks' or 'geom'\n    The helper calls city.add_building with the right kwargs."""
    for e in entries:
        btype = e.get('type')
        door = e.get('door')
        if 'geom' in e:
            city.add_building(btype, door, geom=e['geom'])
        else:
            kwargs = {}
            if 'blocks' in e:
                kwargs['blocks'] = e['blocks']
            city.add_building(btype, door, **kwargs)

# Add a park
_add_many(city, [
    {'type': 'park', 'door': (13, 11), 'geom': box(9, 9, 13, 13)},
])

# Add homes
_add_many(city, homes)

# Add workplaces
_add_many(city, works)

# Add retail places
_add_many(city, retails)

city.get_street_graph()
elapsed = time.perf_counter() - t0
print(f"City built and street graph computed in {elapsed:.3f}s; buildings={len(city.buildings_gdf)} streets={len(city.streets_gdf)}")

city.buildings_gdf.set_crs("EPSG:3857", inplace=True)
city.streets_gdf.set_crs("EPSG:3857", inplace=True)

# Persist as GeoPackage
city.save_geopackage('synthetic_pois.gpkg')

# Plot a city
fig, ax = plt.subplots(figsize=(6, 6))
plt.box(on=False)
city.plot_city(ax, doors=True, address=False)
plt.show(block=False)
plt.close(fig)
```
